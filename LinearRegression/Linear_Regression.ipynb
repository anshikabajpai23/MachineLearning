````python
import pandas as pd

df=pd.read_csv("/content/train.csv")
df=df.dropna()
````

````python
a=np.array(df['x'])
b=np.array(df['y'])
````

````python
#Implemented StandardScaler
def standardization(arr):
  mu=np.mean(arr,axis=0)
  std=np.std(arr,axis=0)
  # print(mu)
  # print(std)
  return (arr - mu)/std
````

````python
import numpy as np

class Linear_regression():
  def __init__(self,alpha,x,y,iterations):
    self.alpha=alpha
    self.x=x
    self.y=y
    self.theta=[]
    self.m=len(x)
    self.itr=iterations

  def cost(self,y,y_pred):
    m=len(y)
    j_theta =(np.sum((y_pred-y)**2))*(1/(2*m))
    return j_theta

  def predict(self, x, theta):
    y_pred = np.dot(theta,np.transpose(x))
    return y_pred

  def gradient(self,theta,x,y):
    y_pred=self.predict(x,theta)
    theta_new=theta - self.alpha*(1/self.m )*np.dot(np.transpose(y_pred-y),x)
    return theta_new

  def preprocessing(self, x):
    #Performing standardization and adding 1's for bias
    x_new=x.reshape(len(x),1)
    x_new=standardization(x_new)

    one=np.ones(len(x)).reshape(len(x),1)
    x_new=np.concatenate((one,x_new),axis=1)

    return x_new

  def fit(self):
    x_train=self.preprocessing(self.x)

    self.theta=np.zeros(len(self.x.shape)+1)
    for i in range(self.itr):
      self.theta = self.gradient(self.theta,x_train,self.y)
      # print(self.theta)
      # print("------next iteration--------")
    return self.theta
````

````python
lr=Linear_regression(0.01,a,b,1)
weights=lr.fit()
a_train=lr.preprocessing(a)
y_pred=lr.predict(a_train,weights)
print("after 1 itr: ",lr.cost(b,y_pred) )

lr=Linear_regression(0.01,a,b,5)
weights=lr.fit()
a_train=lr.preprocessing(a)
y_pred=lr.predict(a_train,weights)
print("after 5 itr: ",lr.cost(b,y_pred) )


lr=Linear_regression(0.01,a,b,100)
weights=lr.fit()
a_train=lr.preprocessing(a)
y_pred=lr.predict(a_train,weights)
print("after 100 itr: ",lr.cost(b,y_pred) )


lr=Linear_regression(0.01,a,b,1000)
weights=lr.fit()
a_train=lr.preprocessing(a)
y_pred=lr.predict(a_train,weights)
print("after 1000 itr: ",lr.cost(b,y_pred) )
````

**Key Things to note:**
<br/>


*   Do not standardize the bias parameter (add ones after standardization)
*   Always perform standardization
*   Make sure the shape is correct for X and Y.
*.  For calculating the gradient, it is always predictions - actual.
